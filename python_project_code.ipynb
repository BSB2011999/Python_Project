```python
import pandas as pd
```

```python
raw_df = pd.read_csv('Reporting_Sample_Dataset.csv')
raw_df
```

```python
df = raw_df.copy()
df = df.groupby('publisher_platform')
df = df.get_group('facebook')
df.head()
```

```python
df.isnull().sum()    
```

```python
fbdf = pd.DataFrame()
fbdf['ad_date']=df['ad_date'].unique()
fbdf
```

```python
df.info()    # ad_date is object, convert it to date time to do sum of sale and other sum operations on date
```

```python
df['ad_date']=pd.to_datetime(df['ad_date'])    # converts date to date time
```

```python
df.info()
```

```python
fbdf['total_spends'] = list(df.resample('D',on='ad_date').spend.sum())
fbdf['total_impressions'] = list(df.resample('D',on='ad_date').impressions.sum())
fbdf['total_clicks'] = list(df.resample('D',on='ad_date').clicks.sum())
fbdf['total_unique_clicks'] = list(df.resample('D',on='ad_date').unique_clicks.sum())
fbdf['cpm'] = fbdf['total_spends'].mul(1000).div(fbdf['total_impressions'])
fbdf['ctr'] = fbdf['total_clicks'].div(fbdf['total_impressions']).mul(100)
fbdf['unique ctr'] = fbdf['total_unique_clicks'].div(fbdf['total_impressions']).mul(100)

```

```python
fbdf
```

```python
platform = []
for i in range(len(fbdf['ad_date'])):
    platform.append('facebook')
print(platform)
```

```python
fbdf['platform']=platform
fbdf.set_index('platform',inplace=True)
fbdf
```

```python
fbdf.to_csv('fbdf.csv')
```
